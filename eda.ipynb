```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Mobile Reviews Sentiment Cleaned.csv')

print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
df.describe()
```

**Rating**

```python
plt.figure(figsize=(6,4))
df['rating'].value_counts().sort_index().plot(kind='bar', color='steelblue')
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

**Sentiment**

```python
plt.figure(figsize=(5,4))
df['sentiment'].value_counts().plot(kind='bar', color=['green','orange','red'])
plt.title('Sentiment Distribution')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='sentiment', y='rating', data=df, order=['Positive','Neutral','Negative'])
plt.title('Rating by Sentiment')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
df['brand'].value_counts().plot(kind='bar', color='coral')
plt.title('Reviews per Brand')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
df.groupby('brand')['price_usd'].median().sort_values().plot(kind='bar', color='teal')
plt.title('Median Price by Brand')
plt.tight_layout()
plt.show()
```

```python
df['source'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(5,5))
plt.title('Review Source')
plt.ylabel('')
plt.show()
```

```python
num_cols = df.select_dtypes(include='number').columns
plt.figure(figsize=(10,8))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
